```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL of the webpage to scrape
url = "https://hoopshype.com/salaries/players/2022-2023/"

# Send an HTTP GET request to the URL
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    print('Request Succesful!')
    # Parse the HTML content of the page using BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Find the table containing the data you want to scrape
    table = soup.find('table', class_='hh-salaries-ranking-table')

    # Initialize empty lists to store the data
    players = []
    salary_2022_23 = []

    # Iterate through the rows of the table and extract the data
    for row in table.find_all('tr')[1:]:  # Skip the header row
        columns = row.find_all('td')
        player = columns[1].text.strip()
        salary = columns[2].text.strip()
        
        players.append(player)
        salary_2022_23.append(salary)

    # Create a Pandas DataFrame from the extracted data
    df = pd.DataFrame({'Player Name': players, 'Salary': salary_2022_23})

    # Print the DataFrame
    print(df)

else:
    print("Failed to retrieve the webpage. Status code:", response.status_code)
```

```python
df2 = pd.read_csv('nba_2022-23_all_stats.csv', index_col=0)

merged_df = pd.merge(df, df2, on='Player Name', how='inner')

merged_df['Salary'] = pd.to_numeric(merged_df['Salary'].str.replace('$', '').str.replace(',', ''), errors='coerce')

merged_df.to_csv('nba_2022-23_all_stats_with_salary.csv')

merged_df.head()
```
